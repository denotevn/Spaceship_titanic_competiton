```python

def deck_side_encode(data):
    split_cabin = pd.DataFrame(list(map(lambda x: str(x).split("/"), data.Cabin)))
    new_data = data.copy()
    new_data["Side"] = split_cabin[2]
    new_data["Deck"] = split_cabin[0]
    new_data = new_data.drop("Cabin", axis=1)
    return new_data
```

### Fill Nan's with means and modes

```python
train = pd.read_csv("/home/tuandinh/Desktop/ML and AI/Spaceship Titanic competiton/datasets/train.csv")
test = pd.read_csv("/home/tuandinh/Desktop/ML and AI/Spaceship Titanic competiton/datasets/test.csv")
train.isnull().sum()
```

```python
train.head()
```

```python
def data_prep(tr, te, drop=None):
    tr_no_na = tr.dropna()
    attribs_drop = ["Name", "PassengerId"]
    train_1 = tr.drop(attribs_drop, axis=1)
    test_1 = te.drop(attribs_drop, axis=1)
    value = {
        "CryoSleep": tr_no_na.CryoSleep.mode()[0],
        "Age" : tr_no_na.Age.mean(),
        "VIP" : tr_no_na.VIP.mode()[0],
        "RoomService" : tr_no_na.RoomService.mean(),
        "FoodCourt" : tr_no_na.FoodCourt.mean(),
        "ShoppingMall" : tr_no_na.ShoppingMall.mean(),
        "Spa" : tr_no_na.Spa.mean(),
        "VRDeck" : tr_no_na.VRDeck.mean()
    }
    train_prep = train_1.fillna(value=value)
    test_prep = test_1.fillna(value=value)
    train_prep["Transported"] = train_prep["Transported"].astype(int)
    train_prep["CryoSleep"] = train_prep["CryoSleep"].astype(int)
    train_prep["VIP"] = train_prep["VIP"].astype(int)
    test_prep["CryoSleep"] = test_prep["CryoSleep"].astype(int)
    test_prep["VIP"] = test_prep["VIP"].astype(int)
    train_prep = pd.get_dummies(deck_side_encode(train_prep), dummy_na=True, columns=["HomePlanet", "Destination", "Side", "Deck"])
    test_prep = pd.get_dummies(deck_side_encode(test_prep), dummy_na=True, columns=["HomePlanet", "Destination", "Side", "Deck"])
    return train_prep, test_prep
```

```python
train_prep,  test_prep = data_prep(train, test)
```
